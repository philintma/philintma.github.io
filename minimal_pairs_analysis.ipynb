```python
from bs4 import BeautifulSoup
import requests
```

```python
pronounciations = []
for letter in range (65, 90): #A to Z
    url = "https://diksiyonaryo.ph/list/" + chr(letter)
    soup = BeautifulSoup(requests.get(url).text, 'html.parser')
    pronounciations.extend([p.text for p in soup.find_all(class_ ="pronunciation")])
# soup
```

```python
pronounciations
```

```python
def min_pair(word1, word2):
    res = False
    if (len(word1) == len(word2) and len(word1) > 2):
        countDif = 0
        letter = False
        for i in range(len(word1)):
            if (word1[i] != word2[i]):
                countDif += 1
                if (word1[i] == "v" or word2[i] == "v"):
                    letter = True
        if (countDif == 1 and letter):
            res = True
    return res
```

```python
min_pairs = set()
for i in range (len(pronounciations)):
    for j in range(i + 1, len(pronounciations)):
        if min_pair(pronounciations[i], pronounciations[j]):
            min_pairs.add((pronounciations[i], pronounciations[j]))

min_pairs
```

```python
f = open("min_pairs.txt", "w")
for s in min_pairs:
    f.write(s[0] + " -- " + s[1] + "\n")
f.close()
```
